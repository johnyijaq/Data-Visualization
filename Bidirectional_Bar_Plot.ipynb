```python
df = pd.read_csv('Bidirectional_Bar_Chart.csv')
df.head()
```

```python
# Set up data for plotting
data = df[['Gene Ontology (Biological  Process) ','Male ','Female']]
data.set_index('Gene Ontology (Biological  Process) ', inplace=True)
```

```python
data['Male '] = data['Male ']
data['Female'] = data['Female']
data['Female'] = data['Female'] * (-1) # Has to multiply with -1 to get the negative values to reverse the order in the x-axis
data
```

```python
#list of variables
font_color = '#525252'
hfont = {'fontname':'sans-serif'}
facecolor = '#eaeaf2' #background color
color_red = '#fd625e'
color_blue = '#01b8aa'
index = data.index
column0 = data['Female']
column1 = data['Male ']
title0 = 'Female'
title1 = 'Male'
```

```python
# Create a new figure and axis
fig, axes = plt.subplots(figsize=(10,5), facecolor = facecolor, ncols=2, sharey=True)
fig.tight_layout()

# Plot the bars and set titles
axes[0].barh(index, column0, align='center', color=color_red, zorder=10)
axes[0].set_title(title0, fontsize=12, pad=15, color=color_red, **hfont)
axes[1].barh(index, column1, align='center', color=color_blue, zorder=10)
axes[1].set_title(title1, fontsize=12, pad=15, color=color_blue, **hfont)

# To show data from highest to lowest
plt.gca().invert_yaxis()


# Add labels/Edit ticks
axes[0].set(yticks=data.index, yticklabels=data.index)
axes[0].yaxis.tick_left()
# Make y-axis labels smaller and left-aligned
for label in axes[0].get_yticklabels():
    label.set_horizontalalignment('left') #Aligns the labels to the left.
    label.set_position((-0.4, label.get_position()[1])) # Moves the labels further to the left. Adjust -0.3 as necessary to avoid overlap with the plot.
axes[0].tick_params(axis='y', colors='white')  #To make the y-axis labels smaller, you can add tick_params to your existing code.axes[0].tick_params(axis='y', labelsize=10)
axes[1].set_xticks([10, 20, 30, 40, 50, 60, 70])
axes[1].set_xticklabels([10, 20, 30, 40, 50, 60, 70])


#Edit tick labels
for label in (axes[0].get_xticklabels() + axes[0].get_yticklabels()):
    label.set(fontsize=8, color=font_color, **hfont)
for label in (axes[1].get_xticklabels() + axes[1].get_yticklabels()):
    label.set(fontsize=8, color=font_color, **hfont)


# Adjust subplot layout to provide more space for y-axis labels
plt.subplots_adjust(wspace=0, top=0.90, bottom=0.1, left=0.18, right=0.95)

#Annotate bars with the number of proteins
for i in data.index:
   # Ensure the text for the left subplot is placed correctly
    axes[0].text(column0[i] - 3, i, str(column0[i]* (-1)), va='center', color='black', fontsize=8) #places the text slightly to the right of the bar's end.
    axes[1].text(column1[i] + 0.5, i, str(column1[i]), va='center', color='black', fontsize=8)

    #y_pos[i] - sets the vertical position of the text to the center of the bar.
    #str(male_values[i]) converts the value to a string to be displayed.
    #va='center' aligns the text vertically to the center of the bar.
    #This adds text annotations to the left subplot (axes[0]).
    #column0[i] + 0.5 places the text slightly to the right of the end of the bar for the male values.
    #i is the index position for the text (since i iterates over data.index).
```
