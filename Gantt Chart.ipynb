```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tasks = {
    "Task": [
        "Literature Review and Biomarker Selection",
        "Peptide Design and Synthesis",
        "Method Development and Optimization",
        "Sample Collection and Preparation",
        "Biomarker Detection and Quantification",
        "Data Analysis and Validation",
        "Final Reporting"
    ],
    "Start Month": [1, 4, 7, 13, 16, 21, 23],
    "End Month": [3, 6, 12, 15, 20, 22, 24]
}
# Define the work plan data with correct month range
```

```python
# Convert to DataFrame
df = pd.DataFrame(tasks)
```

```python
# Define colors for each task
colors = ['skyblue', 'lightgreen', 'lightcoral', 'pink', 'yellow', 'gray', 'lightblue']

```

```python
# Adjusting Gantt chart for a 24-month timeline with different colors
plt.figure(figsize=(10, 6))
for i, task in enumerate(df["Task"]):
    plt.barh(task, df["End Month"][i] - df["Start Month"][i] + 1, left=df["Start Month"][i] - 1, color=colors[i])
# Add labels and title
plt.xlabel('Months')
plt.xticks(range(1, 25))  # Adjusting x-axis to show only 24 months
plt.ylabel('Tasks')
plt.title('Gantt Chart for Project Work Plan (24 Months)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
```
